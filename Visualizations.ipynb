```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('final_df.csv', index_col=[0])
```

```python
df
```

```python
slope, _int, r, p, std_err = stats.linregress(df['weeks_on_list'], df['pages'])
```

```python
fit = slope * df['weeks_on_list'] + _int
```

```python
plt.figure(figsize=[12,12])
plt.scatter(df['weeks_on_list'], df['pages'])
plt.plot(df['weeks_on_list'], fit)
plt.annotate(f'r^2={round(r**2, 2)}', (125,1500))
plt.ylabel('Pages')
plt.xlabel('Weeks on NYT Bestseller List')
plt.show()
```

```python
new_df = df.loc[df['weeks_on_list'] >= 25]
new_df
```

```python
slope, _int, r, p, std_err = stats.linregress(new_df['weeks_on_list'], new_df['pages'])
```

```python
fit = slope * new_df['weeks_on_list'] + _int
```

```python
plt.figure(figsize=[12,12])
plt.scatter(new_df['weeks_on_list'], new_df['pages'])
plt.plot(new_df['weeks_on_list'], fit)
plt.annotate(f'r^2={round(r**2, 2)}', (125,600))
plt.ylabel('Pages')
plt.xlabel('Weeks on NYT Bestseller List')
plt.show()
```
