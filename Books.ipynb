```python
import pandas as pd
import numpy as np
import requests
from config import g_key, ny_key
import datetime as dt
from datetime import timedelta
import time
```

```python
url = 'https://www.googleapis.com/books/v1/volumes?q=isbn:'
best_url = 'https://api.nytimes.com/svc/books/v3/lists/'
```

```python
date_list = []
start_date = dt.date(2011,3,3)
end_date = dt.date(2022,9,8)
delta = end_date - start_date
for i in range(int((delta.days + 7)/7)):
    day = start_date + timedelta(weeks=i)
    date_list.append(day)
```

```python
print((date_list))
```

```python
new_dates = []
for date in date_list:
    convert = date.strftime('%Y-%m-%d')
    new_dates.append(convert)
new_dates
```

```python
genre = 'combined-print-and-e-book-fiction'
i = 0
for value in new_dates:
    query_url = f'{best_url}{value}/{genre}.json?api-key={ny_key}'
    query = requests.get(query_url).json()
    books = query['results']['books']
    results = query['results']
    titles = []
    isbn = []
    weeks = []
    dagger = []
    author = []
    date = []
    ranks = []
    for book in books:
        title = book['title']
        titles.append(title)
        writer = book['author']
        author.append(writer)
        rank = book['rank']
        ranks.append(rank)
        number = book['primary_isbn13']
        isbn.append(number)
        week = book['weeks_on_list']
        weeks.append(week)
        strike = book['dagger']
        dagger.append(strike)
        date.append(results['published_date'])
    data = {}
    data['title'] = titles
    data['author'] = author
    data['rank'] = ranks
    data['isbn'] = isbn
    data['weeks_on_list'] = weeks
    data['dagger'] = dagger
    data['published_date'] = date
    if i == 0:
        df = pd.DataFrame(data)
        time.sleep(6)
    else:
        df2 = pd.DataFrame(data)
        df = pd.concat([df, df2], sort=False).reset_index(drop=True)
        time.sleep(6)
    print(i)
    i += 1
```

```python
df.to_csv('data.csv')
```
